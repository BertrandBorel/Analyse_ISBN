```python
def rechercher(ISBN, choix):
    isbn = canonical(str(ISBN))
    data = isbnlib.meta(isbn, service=choix)
    
    liste = [data['Authors'],data['Title'],data['Year'],data['Publisher'],data['ISBN-13']]
    return liste
```
