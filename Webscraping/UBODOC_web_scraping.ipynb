### Web Scraping + ISBN dans une liste

```python
# Importations 
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

# page 
nombre = 200 # à augmenter pour récupérer le max de livres
limite = 200
livre = 0
url = 'https://bu-catalogue.univ-brest.fr/cgi-bin/koha/opac-detail.pl?biblionumber='+str(nombre)
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# # test :
# liste_test = []



# Variables globales
liste_EAN = []


import re

while livre < limite :
    try : 
        url = 'https://bu-catalogue.univ-brest.fr/cgi-bin/koha/opac-detail.pl?biblionumber='+str(nombre)
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')

        # récupérer l'EAN 
        EAN =soup.find_all("dd")

        # récupère uniquement les chiffres de l'EAN + ajout à la liste
        for x in tqdm(EAN[2]) :
            liste_EAN.append(int(x))
        
        # Incrémentations (nb livres et pour sortir de la boucle while) :
        livre += 1
        nombre += 1
    

    # En cas d'erreur, le livre est ignoré, on passe au suivant
    except :
#         print("Erreur. ISBN introuvable.")
        livre += 1
        nombre += 1
      
        
# Affichage à la fin du programme (nb de livres et la liste ISBN)    
# print("Nombre de livres :", livre)
# print("Nombre d'ISBN :", len(liste_EAN))
# print(liste_EAN)
```

### Récupérer informations

```python
from isbnlib import canonical, meta
import csv
from tqdm import tqdm

# ## Création du csv 
# header = ['Auteur(s)', 'Titre', 'Edition', 'Date', 'ISBN']
# with open(r'data_ISBN.csv', 'a') as f:
#     writer = csv.writer(f)
#         # write the header
#     writer.writerow(header)
#     print("fichier créé")
    
print("-------------------------------")    
print("Recherche en cours...")    
    
    
    
    
nombre_livres = 0
liste_ref = []
curseur = 0
ISBN = []


# ---------------------------------------------------------------------
# Liste ISBN : 'liste_EAN'
while curseur < len(liste_EAN):
    try :
        for x in tqdm(liste_EAN) :
                isbn = canonical(str(x))
                data = meta(isbn)
                nombre_livres += 1 


                biblio = []

                if len(data["Authors"]) == 1 :
                    for x in data["Authors"] :
                        biblio.append(x)
                else : 
                    biblio.append(data["Authors"])

                biblio.append(data["Title"])
                biblio.append(data["Publisher"])
                biblio.append(data["Year"])
                biblio.append(data['ISBN-13'])


                ISBN.append(biblio)
                curseur += 1

    except :
        print("Erreur ISBN")
        curseur += 1

print("-------------------------------")
print("Nombre de livres dans la liste : ", nombre_livres)
print("-------------------------------")
print(ISBN)
print("-------------------------------")
print("Insertion dans le csv...")
    
with open(r'data_ISBN.csv', 'a', encoding='UTF8') as f:
        for x in tqdm(ISBN) :
            writer = csv.writer(f)
            writer.writerow(x)
    
        print("Insertion des données réussies.")
```
