# <center> Programme : trouver les informations d'un livre grâce à l'isbn </center>

```python
from isbnlib import canonical, meta


# Saisie de l'ISBN
print("Note : L'ISBN d'un ouvrage doit être rentré sans espace et sans tiret de séparation.\n Les chiffres doivent être collés comme l'exemple suivant : 97845678822")
print("-----------------------------------------")
ISBN = input("Entrez l'ISBN de l'ouvrage : ")
print("ISBN saisi :", ISBN)
print("Recherche en cours...")


# Récupération des informations
isbn = canonical(ISBN)
data = meta(isbn)

biblio = []

biblio.append(data["Authors"])
biblio.append(data["Title"])
biblio.append(data["Publisher"])
biblio.append(data["Year"])
biblio.append(data['ISBN-13'])

# Affichage de la réponse
print("-----------------------------------------")
print("Résultat de la recherche : ")
print(biblio)

# Sauvegarde possible :
print("\n")
print("-----------------------------------------")
reponse=input("Désirez-vous faire une sauvegarde du résultat ? (o/n)")
if reponse == "o":
    # sauvegarde
    f = open("save_biblio.txt", "x")
    f.write(str(biblio))
    f.close()
else :
    print("Pas de sauvegarde effectuée.")
```
